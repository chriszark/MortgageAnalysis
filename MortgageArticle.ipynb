```python
import pandas as pd
import matplotlib.pyplot as plot


# Load the XLSX file
varMortgageReview = pd.ExcelFile('MortgageReview.xlsx')

# Create an empty dictionary to store the dataframes
dfs = {}

# Loop through each sheet in the XLSX file
for sheet_name in varMortgageReview.sheet_names:
    # Read the sheet into a dataframe
    df = varMortgageReview.parse(sheet_name)
    
    # Store the dataframe in the dictionary with the sheet name as the key
    dfs[sheet_name] = df


# Create a summary dataframe with the names of all dataframes
summary_df = pd.DataFrame({'Sheet Name': list(dfs.keys()), 'Dataframe': list(dfs.values())})
```

```python
unselected_columns = ['Average Period of mortgage']
df_selected = df.drop(unselected_columns, axis=1)

# Plot the line chart for each category
ax = df_selected.iloc[:, :-2].plot(kind='line', marker='o')


# Set axis labels and legends
ax.set_xlabel('Year')
ax.set_ylabel('%')
ax.legend(title='Category',bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```

```python
dfFixedandVariableRates = dfs['FixedandVariableRates']
selected = ['Date','Variable','Fixed']
dfSelected = dfFixedandVariableRates[selected]
```

```python

# Convert the 'Date' column to datetime and sort the DataFrame by date
dfSelected['Date'] = pd.to_datetime(dfSelected['Date'])
dfSelected.sort_values('Date', inplace=True)

# Set the 'Date' column as the index
dfSelected.set_index('Date', inplace=True)

# Calculate the percentages for each variable
dfSelected_percent = dfSelected.div(dfSelected.sum(axis=1), axis=0) * 100

# Plot the 100% stacked bar chart
ax = dfSelected_percent.plot(kind='bar', stacked=True)

# Set axis labels and legends
ax.set_xlabel('Date')
ax.set_ylabel('Percentage')
ax.legend(title='Variable', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```
